```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import axis as ax
%matplotlib inline
import seaborn as sns
sns.set_style("whitegrid")
```

```python
df=pd.read_csv('Laptop.csv')
df=df.reset_index(drop=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=90)
sns.countplot(x=df['Brand'])
plt.title('number of models for each brand',fontsize=20)
plt.xlabel('Brand',fontsize=15)
plt.ylabel('number of models',fontsize=15)
```

```python
all_models=df['Brand'].value_counts()
all_models_sum=df['Brand'].value_counts().sum()
Dell=all_models[1]
Dell_percent=Dell/all_models_sum * 100
Dell_percent
```

```python
plt.figure(figsize=(15,8))
all_models.plot(kind='pie',fontsize=12)
```

```python
processor=df['Processor']
```

```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=90)
sns.countplot(x=processor)
plt.title("count of processors",fontsize=20)
```

```python
plt.figure(figsize=(15,8))
pro_count=processor.value_counts()
pro_count.plot(kind="pie",fontsize=12)
```

```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=90)
sns.countplot(x=df['Hard_Disk_Capacity'])
```

```python
prices=df['Price']
```

```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=90)
prices.plot(kind='hist')
plt.axvline(prices.mean(), color='r', linestyle='dashed', linewidth=3,label='the mean')
plt.text(prices.mean()*1.025, 25, 'Mean: {:.2f}'.format(prices.mean()),fontsize=20)
```
